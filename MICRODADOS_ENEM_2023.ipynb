```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

# Uma função utilitária para a impressão de títulos
def t (*str, linhas_antes = 2, linhas_depois = 1, primeira_linha_separada = True):
    print('\n'*linhas_antes, end='')
    for s in str:
        print(s)
        if primeira_linha_separada:
            print('')
            primeira_linha_separada = False
    print('\n'*linhas_depois, end='')
```

```python
ARQUIVO_CSV = 'MICRODADOS_ENEM_2023.csv'
enem = pd.read_csv(ARQUIVO_CSV, delimiter=';', decimal='.')
```
